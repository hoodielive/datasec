```python
def jaccard(first, second):
  first = set(first)
  second = set(second)
  return len(first & second) / float(len(first | second))
```

```python
def recommend(given_user, K=5):
    similarities = {}  # create similarities dictionary
    for user_id, brands in brandsfor.iteritems():
        score = jaccard(given_user, brands)
        if score < 1 and score > 0:  # only want people that are not exactly the sme but have at least something in common
            similarities[user_id] = score
        
    most_similar_users = sorted(similarities.items(), key = lambda x: x[1], reverse=True)[:K]
    # this might have people with a similarity of 1.0, which is bad
    print most_similar_users
    
    brands_to_recommend = []
    for user in most_similar_users:
        for brand in brandsfor[user[0]]:
            if brand not in given_user:  # if the brand is not already liked by the given user
                brands_to_recommend.append(brand)
                
    return dict(Counter(brands_to_recommend))

```
